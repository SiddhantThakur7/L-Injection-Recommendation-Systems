```python
#Creating Training and tesing sets
def split_train_test(r):
    test_set = np.zeros(r.shape)
    train_set = r.copy()
    for i in range(r.shape[0]):
        test_indices = np.random.choice(np.flatnonzero(r[i]), size = 10, replace = False)
        
    train_set[i, test_indices] = 0.0
    test_set[i, test_indices] = r[i, test_indices]
    
    return train_set, test_set
```

```python
class Pre_use_Predictions:
    def __init__(self, iters, reg_param, num_features):
        self.iters = iters
        self.reg = reg_param
        self.n_features = num_features
        
    '''def core_function(self, ru, pu, Y, Z, alpha):
        V = np.dot(Y, ru)
        W = np.dot(Y, pu)
        b = np.dot(Y,(1+alpha*(pu)))
        #b = 0
        #j = 0
        A = Z + alpha*(np.dot(V, W.T)) + self.reg*(np.eye(Y.shape[0]))
        c, low = cho_factor(A)
        x = cho_solve((c, low), b)
        return x'''
        
        
    def wALS(self, R, x, y, token):
        #gradient descent of the Objective Function used in Alternating Least Squares
        '''op = R.dot(y)
        M = y.T.dot(y) + (np.eye(self.n_features)*self.reg)
        Minv = np.linalg.inv(M)
        x = op.dot(Minv)
        return x'''
    
        if token == 1:
            for user in range(R.shape[0]):
                Wu = np.diag(C[user])
                a = np.dot(y, np.dot(Wu, R[user]))
                Ainv = np.linalg.inv(np.dot(np.dot(y,Wu), y.T) + np.eye(self.n_features)*(self.reg*np.sum(R[user] > 0)))
                x[:, user] = a.dot(Ainv)
            return x
                
        elif token == 0:    
            for item in range(R.shape[1]):
                Wi = np.diag(C[:, item])
                a = np.dot(x, np.dot(Wi, R[:, item]))
                Ainv = np.linalg.inv(np.dot(np.dot(x, Wi),x.T) + (np.eye(self.n_features)*(self.reg*np.sum(R[:, item] > 0))))
                y[:, item] = a.dot(Ainv)
            return y
        
        '''Z = np.dot(Y, Y.T)
        for u in range(X.shape[1]):
            X[:, u] = self.core_function(R[u], P[u], Y, Z, alpha)
        
        Z = np.dot(X, X.T)
        for i in range(Y.shape[1]):
            Y[:, u] = self.core_function(R[:, i], P[:, i], X, Z, alpha)
        
        return X, Y'''
    
    def prediction(self):
        pred = np.dot(self.X.T, self.Y)
        return pred
    
    def calc_mse(self, M, M2):
        indices = np.nonzero(M)
        mse = mean_squared_error(M[indices], M2[indices])
        return mse
    
    def train_model(self, train, test):
        self.n_user , self.n_movies = train.shape
        
        #initialization of user and item matrices before model is trained 
        self.X = np.random.randn(self.n_features, self.n_user)
        self.Y = np.random.randn(self.n_features, self.n_movies)
        
        self.test_mse = []
        self.train_mse = []
        pool = mp.Pool(mp.cpu_count())
        
        for q in range(self.iters):
            self.X = self.wALS(train, self.X, self.Y, token = 1)
            self.Y = self.wALS(train, self.X, self.Y, token = 0)
            p = self.prediction()
            self.test_mse.append(self.calc_mse(test, p))
            self.train_mse.append(self.calc_mse(train, p))
            
        return p
    @staticmethod
    def plot_learning_curve(model):
        linewidth = 3
        plt.plot(model.test_mse, label = 'Test', linewidth = linewidth)
        plt.plot(model.train_mse, label = 'Train', linewidth = linewidth)
        plt.xlabel('num of iterations')
        plt.ylabel('MSE val')
        plt.legend(loc = 'best')
            
            

```

```python
data = Svd_train

algo = SVD(n_epochs= 20, lr_all= 0.01, reg_all= 0.01)

cross_validate(algo, data, measures=['RMSE', 'MAE'], cv = 5, verbose = True)
```

```python
# test set is made of 25% of the ratings.
trainset, testset = train_test_split(Svd_test, test_size=1.0)
trainset = data.build_full_trainset()
# We'll use the famous SVD algorithm.
algo = SVD(n_epochs= 20, lr_all= 0.01, reg_all= 0.01)

# Train the algorithm on the trainset, and predict ratings for the testset
algo.fit(trainset)
predictions = algo.test(testset)
predictions
```

```python
def get_Iu(uid):
    """ return the number of items rated by given user
    args: 
      uid: the id of the user
    returns: 
      the number of items rated by the user
    """
    try:
        return len(trainset.ur[trainset.to_inner_uid(uid)])
    except ValueError: # user was not part of the trainset
        return 0
    
def get_Ui(iid):
    """ return number of users that have rated given item
    args:
      iid: the raw id of the item
    returns:
      the number of users that have rated the item.
    """
    try: 
        return len(trainset.ir[trainset.to_inner_iid(iid)])
    except ValueError:
        return 0

df = pd.DataFrame(predictions, columns=['uid', 'iid', 'rui', 'est', 'details'])
df['Iu'] = df.uid.apply(get_Iu)
df['Ui'] = df.iid.apply(get_Ui)
df['err'] = np.square(df.est - df.rui)

##Best and Worst Predictions
best_predictions = df[df['uid'] == 1].sort_values(by='err')[-1:-10:-1]
worst_predictions = df.sort_values(by='err')[-10:]
best_predictions
```
