# Creation Of Pre-Use_Preference Matrix
First, we build a pre-use preference matrix by examining a rating matrix R. It is set as one if rui 2 R has been already rated (i.e., u should have liked i if she bought i)
1. (Step 1). It is the highest because pui is set as a real value in {0, 1}. 
2. (Step 2)., we infer pre-use preference scores on “unrated” user-item pairs; (i.e., puj null) based on other observed pre-use preferences (i.e., pui 1) and add them in P, which becomes P^

# Creating a helper function for splitting the data

Since, this is a recommendation data we cant split simply using a split Ratio\
**Since, we need to provide recommendations to each user available we cant afford to have a users interaction in test set and not in the training set.**

Hence we do the following:
1. As we know from above each user has rated atleast 18 items.**We choose at random 10 out of those atleast 18 movies.**

2. We transfer those ratings to the test set. Now these items behave as items rated by corresponding user.\

3. The items transfered to the test set now behave as items not rated by corresponding user.



```python
#Creating Training and tesing sets
def split_train_test(r):
    test_set = np.zeros(r.shape)
    train_set = r.copy()
    for i in range(r.shape[0]):
        test_indices = np.random.choice(np.flatnonzero(r[i]), size = 10, replace = False)
        
    train_set[i, test_indices] = 0.0
    test_set[i, test_indices] = r[i, test_indices]
    
    return train_set, test_set
```

# Weight Sampling for Weighted Alternating Least Squares

## **Schemes**
1. **Uniform Scheme**\
Each Unrated item has equal probability of being liked and disliked for each user irrespective to his or her pior interaction. Pbij ∝ 1

With increasing α reflecting more confidence


2. **User-Oriented Scheme**
The underlying principle of the useroriented scheme is essentially that as a user rates more items, she is more likely to dislike unrated items.\
it computes the weight wui in proportion to the number of items rated by u:



3. **Item-Oriented Scheme**
Means that if an item is viewed by less users, those users that have not viewed the item will never view it either.

# Weighted Alternating Least Squares
The OCCF method finally updates pui 2 P using their corresponding weights. We treat the updated values as the inferred pre-use preference scores. To update these values, the OCCF method employs the weighted alternating least squares (wALS) method.

The wALS method decomposes a matrix P into two low-rank matrices X and Y while optimizing an objective function. The matrix P represents observed pre-use preferences in our case, i.e., P. The matrices X and Y represent the latent features of users and items, respectively. 
1. **The objective function**


where pui and wui are the entries in the observed pre-use preference matrix P and its weight matrix W, respectively. The vector Xu is the uth row of matrix X, and the vector Yi is the ith row of matrix Y . The two vectors represent the features of user u and item i.

2. Gradient with respect to X


3. Gradient with respect to Y




**In order to factorize matrix P, the OCCF method first assigns random values to elements in matrix Y , and updates elements in matrix X./
Next, the OCCF method updates elements in matrix Y while fixing matrix X**

**Hence due to its alternating nature its called as Alternating Least Squares algorithm.**



```python
class Pre_use_Predictions:
    def __init__(self, iters, reg_param, num_features):
        self.iters = iters
        self.reg = reg_param
        self.n_features = num_features
        
    '''def core_function(self, ru, pu, Y, Z, alpha):
        V = np.dot(Y, ru)
        W = np.dot(Y, pu)
        b = np.dot(Y,(1+alpha*(pu)))
        #b = 0
        #j = 0
        A = Z + alpha*(np.dot(V, W.T)) + self.reg*(np.eye(Y.shape[0]))
        c, low = cho_factor(A)
        x = cho_solve((c, low), b)
        return x'''
        
        
    def wALS(self, R, x, y, token):
        #gradient descent of the Objective Function used in Alternating Least Squares
        '''op = R.dot(y)
        M = y.T.dot(y) + (np.eye(self.n_features)*self.reg)
        Minv = np.linalg.inv(M)
        x = op.dot(Minv)
        return x'''
    
        if token == 1:
            for user in range(R.shape[0]):
                Wu = np.diag(C[user])
                a = np.dot(y, np.dot(Wu, R[user]))
                Ainv = np.linalg.inv(np.dot(np.dot(y,Wu), y.T) + np.eye(self.n_features)*(self.reg*np.sum(R[user] > 0)))
                x[:, user] = a.dot(Ainv)
            return x
                
        elif token == 0:    
            for item in range(R.shape[1]):
                Wi = np.diag(C[:, item])
                a = np.dot(x, np.dot(Wi, R[:, item]))
                Ainv = np.linalg.inv(np.dot(np.dot(x, Wi),x.T) + (np.eye(self.n_features)*(self.reg*np.sum(R[:, item] > 0))))
                y[:, item] = a.dot(Ainv)
            return y
        
        '''Z = np.dot(Y, Y.T)
        for u in range(X.shape[1]):
            X[:, u] = self.core_function(R[u], P[u], Y, Z, alpha)
        
        Z = np.dot(X, X.T)
        for i in range(Y.shape[1]):
            Y[:, u] = self.core_function(R[:, i], P[:, i], X, Z, alpha)
        
        return X, Y'''
    
    def prediction(self):
        pred = np.dot(self.X.T, self.Y)
        return pred
    
    def calc_mse(self, M, M2):
        indices = np.nonzero(M)
        mse = mean_squared_error(M[indices], M2[indices])
        return mse
    
    def train_model(self, train, test):
        self.n_user , self.n_movies = train.shape
        
        #initialization of user and item matrices before model is trained 
        self.X = np.random.randn(self.n_features, self.n_user)
        self.Y = np.random.randn(self.n_features, self.n_movies)
        
        self.test_mse = []
        self.train_mse = []
        pool = mp.Pool(mp.cpu_count())
        
        for q in range(self.iters):
            self.X = self.wALS(train, self.X, self.Y, token = 1)
            self.Y = self.wALS(train, self.X, self.Y, token = 0)
            p = self.prediction()
            self.test_mse.append(self.calc_mse(test, p))
            self.train_mse.append(self.calc_mse(train, p))
            
        return p
    @staticmethod
    def plot_learning_curve(model):
        linewidth = 3
        plt.plot(model.test_mse, label = 'Test', linewidth = linewidth)
        plt.plot(model.train_mse, label = 'Train', linewidth = linewidth)
        plt.xlabel('num of iterations')
        plt.ylabel('MSE val')
        plt.legend(loc = 'best')
            
            

```

## Infering Pre-Use-Preferences of Unrated Items
Based on the pre-use preference scores inferred by the OCCF method, the uninteresting items of user u are defined as follows:


If **theta** is set high, a large number of unrated entries are injected with low values, leading to a less sparse rating matrix. On the other hand, if u is set low, we may not be fully utilizing the benefit of uninteresting items as only a small number of unrated entries are injected. The simple use of relative cut-off based on the percentile rank works well.

Hence **theta** is a Hyperparameter and needs to be tuned.


L-injection is such that we assign a “low” value to rui in R if an item i is determined as uninteresting for a user u. This is because u would not be satisfied with an uninteresting item i even if recommended.

By filling a rating matrix R with low values, we can build a new “denser” matrix that contains low value ratings as well as actual user ratings. We call this augmented matrix an l-injected matrix L.


The Process of L-injection is as follows:


We defile Vui as follows:



Where **delta** is the Scaling factor with respect to averages and needs to be tuned.



```python
def Create_Training_set(arr):
    test = arr
    users = []
    items = []
    stars = []
    for i in range(arr.shape[0]):
        a = np.where(test[i] > 0)[0]
        stars.extend(np.take(test[i], a))
        items.extend([k+1 for k in list(a)])
        U = [i+1]*len(a)
        users.extend(U)

    star_dict= {'userID':users,
                'itemsID':items,
                'ratings':stars
                }
    Ratings_for_Svd=pd.DataFrame(star_dict)
    return Ratings_for_Svd
#Create_Training_set(np.array(Aug_df))
```

```python
def Create_test_Set(arr):
    test = arr
    users = []
    items = []
    stars = []
    for i in range(arr.shape[0]):
        a = np.where(test[i] == 0)[0]
        stars.extend(np.take(test[i], a))
        items.extend([k+1 for k in list(a)])
        U = [i+1]*len(a)
        users.extend(U)


    star_dict= {'userID':users,
                'itemsID':items,
                'ratings':stars
                }
    Recommendations_for_Svd=pd.DataFrame(star_dict)
    return Recommendations_for_Svd
#Recommendations_for_Svd = Create_test_Set(np.array(Aug_df))
```

SVD states that any matrix A can be factorized as:

where U and V are orthogonal matrices with orthonormal eigenvectors chosen from AAᵀ and AᵀA respectively. S is a diagonal matrix with r elements equal to the root of the positive eigenvalues of AAᵀ or Aᵀ A (both matrics have the same positive eigenvalues anyway). The diagonal elements are composed of singular values.


If user u is unknown, then the bias bu and the factors pu are assumed to be zero. The same applies for item i with bi and qi.

The minimization is performed by a very straightforward stochastic gradient descent.

# NMF
NMF **(Nonnegative Matrix Factorization)**  is a matrix factorization method where we constrain the matrices to be nonnegative.

NMF is suited for tasks where the underlying factors can be interpreted as non-negative.

**The Objective Function**


**Multiplicative Update** --> In this method, W  and H   are each updated iteratively according to the following rule:


 
Rescaled gradient descent can be written as follows:


SVD helps in giving Eigen vectors of the input matrix. The technique is used generally where Eigen Vectors are of interest to us. PCA is one classic example.

NMF is another decomposition technique where we try to split the matrix R into the product of two matrices U and V. Our objective here is to get the two matrices U and V, so that their matrix multiplication gives us the matrix R. Here R is typically very very sparse, which is true in a lot of recommender systems. In such cases NMF works better as the missing-values assumption is inbuilt to the algo.

In case of SVD, it doesn’t assume anything about missing values. So you need to give some missing value imputation for SVD. This might bring in unnecessary noise. But if your ratings matrix is not too sparse, SVD might produce better results. SVD results are more deterministic compared to that of NMF. That also can play a role in which method you choose.

# Comparitive Evaluation of all the applied models

Comparison with respect to:

1. **Root Mean Square Error**\
Root Mean Square Error (RMSE) is the standard deviation of the residuals (prediction errors).


2. **Mean Absolute Error**\
Mean Absolute Error (MAE): This measures the absolute average distance between the real data and the predicted data, but it fails to punish large errors in prediction.


3. **FCP (Fraction of Concordant Pairs)**\
Given a test set Rˆ, we define the number of concordant
pairs for user u by counting those ranked correctly by rating
predictor rˆu·

Summing over all users we define nc:

nd = 

. The quality metric we use measures the proportion
of well ranked items pairs, denoted by **FCP (for Fraction of
Concordant Pairs)**


a measure that generalizes the known **AUC** metric into nonbinary ordered outcomes.

```python
benchmark = [{'Algorithm': 'SVD++', 'test_rmse': 0.45986937, 'test_mae': 0.212291, 'test_fcp': 0.734317,'fit_time': 10504.443662643433,'test_time': 147.40530061721802}]
data = Svd_train
# Iterate over all algorithms
for algorithm in [KNNBasic(k=20, min_k=5, sim_options={'name': 'cosine', 'user_based': False}, verbose=False),
                  KNNWithMeans(k=15, min_k=3, sim_options={'name': 'cosine', 'user_based': False}, verbose=False),
                  KNNWithZScore(k=20, min_k=5, sim_options={'name': 'cosine', 'user_based': False}, verbose=False),
                  SVD(n_epochs= 40, lr_all= 0.005, reg_all= 0.01),
                  #SVDpp(n_factors = 10, n_epochs = 40, lr_all = 0.005, reg_all = 0.01),
                  NMF(n_factors = 10, biased = False, init_high= 2)]:
    # Perform cross validation
    results = cross_validate(algorithm, data, measures=['rmse','mae','fcp'], cv=5, verbose=False)
    
    # Get results & append algorithm name
    tmp = pd.DataFrame.from_dict(results).mean(axis=0)
    tmp = tmp.append(pd.Series([str(algorithm).split(' ')[0].split('.')[-1]], index=['Algorithm']))
    benchmark.append(tmp)
    print('Algorithm: {} ... Done'.format(algorithm))
```
